```python
tea['label']=tea['Month'].astype(str)+ '('+ tea['Region'].astype(str)+"-" +round(tea['Product'],2).astype(str)+ ","+round(tea['Price'],2).astype(str)+')'

```

```python
tea= tea.rename(columns={'product ': 'Product'})
```

```python
tea.columns = tea.columns.str.strip()
```

```python
tea_label=tea['label']

tea.head()
```

```python
tea.drop(['label'], axis=1, inplace=True)
```

```python
tea.head()
```

```python
import sklearn
from sklearn.preprocessing import StandardScaler
standardizer= StandardScaler()
standardizer.fit(tea)
std_X=standardizer.transform(tea)
std_tea=pd.DataFrame(std_X, columns=tea.columns)
```

```python
std_tea.describe()
```

```python
from scipy.cluster .hierarchy import linkage, dendrogram
linkage_matrix=linkage(std_tea, method='average', metric='euclidean')
import matplotlib.pyplot as plt
%matplotlib notebook
fig, axs=plt.subplots()
dendrogram(linkage_matrix, labels=tea_label.to_numpy())
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_clust=AgglomerativeClustering(n_clusters=6, affinity='euclidean', linkage='ward')
agg_clusters=agg_clust.fit_predict(std_tea)
agg_result=pd.DataFrame({"label":tea_label, "agg_cluster":agg_clusters}).sort_values('agg_cluster')
agg_result.head()
```
